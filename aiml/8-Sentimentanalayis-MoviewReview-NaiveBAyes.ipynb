# Back to real data: movie reviews

```python
# Split data into training and test sets
from sklearn.model_selection import train_test_split
docs_train, docs_test, y_train, y_test = train_test_split(movie.data, movie.target, 
                                                          test_size = 0.20, random_state = 12)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
movie_tfidVectorizer = TfidfVectorizer(stop_words = 'english', max_df=0.5, ngram_range = (1,1), max_features=5000)
docs_train_tfidVectors = movie_tfidVectorizer.fit_transform(docs_train)
print(docs_train_tfidVectors.toarray())

print(movie_tfidVectorizer.get_feature_names())
```

```python
print(movie_tfidVectorizer.vocabulary_)
print(movie_tfidVectorizer.idf_)
print(movie_tfidVectorizer.stop_words_)
```

# The feature extraction functions and traning data are ready.
Vectorizer and transformer have been built from the training data
Training data text was also turned into TF-IDF vector form
Next up: test data
You have to prepare the test data using the same feature extraction scheme

```python
# Using the fitted vectorizer and transformer, tranform the test data
docs_test_tfidfvectors = movie_tfidVectorizer.transform(docs_test)
```

# Training and testing a Naive Bayes classifier

```python
# Now ready to build a classifier. 
# We will use Multinominal Naive Bayes as our model
from sklearn.naive_bayes import MultinomialNB
```

```python
# Train a Multi Naive Bayes classifier. Again, we call it "fitting"
clf = MultinomialNB()
clf.fit(docs_train_tfidVectors, y_train)
```

```python
# Predict the Test set results, find accuracy
y_pred = clf.predict(docs_test_tfidfvectors)
sklearn.metrics.accuracy_score(y_test, y_pred)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Train a Gaussian Naive Bayes classifier. Again, we call it "fitting"
from sklearn.naive_bayes import GaussianNB
clfG = GaussianNB()
clfG.fit(docs_train_tfidVectors.toarray(), y_train)
# Predict the Test set results, find accuracy
y_pred = clfG.predict(docs_test_tfidfvectors.toarray())
print(sklearn.metrics.accuracy_score(y_test, y_pred))

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
